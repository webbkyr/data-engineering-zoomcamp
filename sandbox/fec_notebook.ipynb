```python
import pandas as pd
# https://www.fec.gov/campaign-finance-data/current-campaigns-house-and-senate-file-description/
# 2026 midterm cycle = 2025-present
column_names = [
    'cand_id',
    'cand_name',
    'cand_ici', # C=challenger, I=Incumbent, O=open seat
    'pty_cd',
    'cand_pty_affiliation',
    'total_receipts',
    'trans_from_auth', 
    'total_disb',
    'trans_to_auth',
    'coh_bop',
    'coh_cop',
    'cand_contrib',
    'cand_loans',
    'other_loans',
    'cand_loan_repay',
    'other_loan_repay',
    'debts_owed_by',
    'total_indiv_contrib',
    'cand_office_st',
    'cand_office_district',
    'spec_election',
    'prim_election',
    'run_election',
    'gen_election',
    'gen_election_percent',
    'other_pol_cmte_contrib',
    'pol_pty_contrib',
    'coverage_end_dt',
    'indiv_refunds',
    'cmte_refunds'
]
dtype_mapping = {
    'cand_id': 'string',
    'cand_name': 'string',
    'cand_ici': 'string', # C=challenger, I=Incumbent, O=open seat
    'pty_cd': 'string',
    'cand_pty_affiliation': 'string',
    'total_receipts': 'float64',
    'trans_from_auth': 'float64',
    'total_disb': 'float64',
    'trans_to_auth': 'float64',
    'coh_bop': 'float64',
    'coh_cop': 'float64',
    'cand_contrib': 'float64',
    'cand_loans': 'float64',
    'other_loans': 'float64',
    'cand_loan_repay': 'float64',
    'other_loan_repay': 'float64',
    'debts_owed_by': 'float64',
    'total_indiv_contrib': 'float64',
    'cand_office_st': 'string',
    'cand_office_district': 'string',
    'spec_election': 'string',
    'prim_election': 'string',
    'run_election': 'string',
    'gen_election': 'string',
    'gen_election_percent': 'float64',
    'other_pol_cmte_contrib': 'float64',
    'pol_pty_contrib': 'float64',
    'indiv_refunds': 'float64',
    'cmte_refunds': 'float64'
}
dtype_dates = ['coverage_end_dt']
# House/Senate current campaigns
cand_df = pd.read_csv(
    './data/weball26.zip',
    compression='zip',
    sep='|',
    dtype=dtype_mapping,
    parse_dates=dtype_dates,
    header=None,
    names=column_names
)
```
